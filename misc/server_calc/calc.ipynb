```python
import socket

host = '2048.challs.olicyber.it'
port = 10007

s = socket.create_connection((host, port))

data = s.recv(2048).decode()

counter = 0

# Parse the operation
while True:
  lines = data.strip().split('\n')
  if len(lines) > 1:
    op_line = lines[-1]
  else:
    op_line = lines[0]

  parts = op_line.split()
  if len(parts) == 3:
    operation = parts[0]
    num1 = int(parts[1])
    num2 = int(parts[2])

    counter += 1

    print(f"Operation {counter}: {operation} {num1} {num2}")
    
    if operation == "SOMMA":
      result = num1 + num2
    elif operation == "DIFFERENZA":
      result = num1 - num2
    elif operation == "PRODOTTO":
      result = num1 * num2
    elif operation == "DIVISIONE":
      result = num1 // num2
    elif operation == "DIVISIONE_INTERA":
      result = num1 // num2
    elif operation == "POTENZA":
      result = num1 ** num2
    else:
      print("Unknown operation:", operation)
      break

    s.sendall(f"{result}\n".encode())
    try:
      data = s.recv(2048).decode()
      if not data:
        print("Connection closed by server.")
        break
    except Exception as e:
      print(f"Error receiving data: {e}")
      break
  else:
    print("Malformed operation line:", op_line)
    data = s.recv(2048).decode()  
    if data == "":
      print("No more data received, exiting.")
      break
```

```python

def xor(a, b):
    return bytes([x^y for x,y in zip(a,b)])

ciphertext = bytes.fromhex('104e137f425954137f74107f525511457f5468134d7f146c4c')

# Try all possible 1-byte keys
for key in range(256):
  key_bytes = bytes([key]) * len(ciphertext)
  plaintext = xor(ciphertext, key_bytes)
  try:
    decoded = plaintext.decode()
    if all(c.isalnum() or c == '_' for c in decoded):
      print(f"Key: {key}, Decoded: {decoded}")
  except UnicodeDecodeError:
    continue
```
